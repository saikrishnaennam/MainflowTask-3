```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('householdtask3.csv')
dataset.head(10)
```

```python
plt.scatter(dataset['year'], dataset['own'])
plt.title('Scatter plot')
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

```python
plt.plot(dataset['year'])
plt.plot(dataset['own'])
plt.title('Line Chart')
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

```python
plt.bar(dataset['year'], dataset['own'])
plt.title('Bar Chart')
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

```python
plt.hist(dataset['income'])
plt.title('histogram')
plt.show()
```
